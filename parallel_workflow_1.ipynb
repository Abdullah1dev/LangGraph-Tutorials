```python
from langgraph.graph import StateGraph , START , END
from typing import TypedDict
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os
```

```python
class Batsman(TypedDict):
    runs: int
    balls: int
    fours: int
    six: int
    
    sr: float
    bpb: float
    boundary_percent: float
    summary: str
    
    
    
```

```python
def calculate_sr(state: Batsman):
    sr = (state['runs'] / state['balls']) * 100
    
    return {'sr' : sr}
```

```python
def calculate_bpb(state: Batsman):
    bpb = (state['balls'] / state['fours'] + state['six']) * 100
    
    
    return {'bpb' : bpb}

```

```python
def calculate_bp(state: Batsman):
    bp = (state['fours'] * 4) + (state['six'] * 6) / state['runs']
    
    return {'bp' : bp}
```

```python
graph = StateGraph(Batsman)

graph.add_node('calculate_sr' , calculate_sr)
graph.add_node('calculate_bp' , calculate_bp)
graph.add_node('calculate_bpb' , calculate_bpb)



graph.add_edge(START , 'calculate_sr')
graph.add_edge(START , 'calculate_bp')
graph.add_edge(START , 'calculate_bpb')

graph.add_edge('calculate_sr' , END)
graph.add_edge('calculate_bp' , END)
graph.add_edge('calculate_bpb' , END)

workflow = graph.compile()

```

```python
initial_state = {
    'runs' : 100,
    'balls' : 60,
    'fours' : 10,
    'six' : 4
    
}
final_state =workflow.invoke(initial_state)
print(final_state)
```
